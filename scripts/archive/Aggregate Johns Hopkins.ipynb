```python
# Dependencies
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
import io
import time
from datetime import datetime, timedelta, date
import json
import pprint
pp = pprint.PrettyPrinter(indent=4)
pd.set_option('display.max_columns', None)

global_reports_url_stem = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/"
```

```python
def getpd(url):
    github_request=requests.get(url).content
    df = pd.read_csv(io.StringIO(github_request.decode('utf-8')))
    return df
```

```python
start_date = date(2020, 1, 22)
current_date = start_date
end_date = date(2020, 10, 22)
delta = timedelta(days=1)
while current_date < end_date:
    file = current_date.strftime("%m-%d-%Y") + ".csv"
    csvURL = global_reports_url_stem + file
    print("Getting " + csvURL)
    if current_date == start_date:
        df = pd.read_csv(csvURL)
        df["Date"] = current_date
    else:
        new_df = pd.read_csv(csvURL)
        new_df["Date"] = current_date
        df = pd.concat([df, new_df], sort=False)
    current_date += delta
df.head()
```

```python
df["Province/State"] = df["Province/State"].astype(str)
df["Province_State"] = df["Province_State"].astype(str)
df["Country/Region"] = df["Country/Region"].astype(str)
df["Country_Region"] = df["Country_Region"].astype(str)
df["Last Update"] = df["Last Update"].astype(str)
df["Last_Update"] = df["Last_Update"].astype(str)
df["State_Province"] = df["Province/State"] + df["Province_State"]
df["Country"] = df["Country/Region"] + df["Country_Region"]
df["Updated"] = df["Last Update"] + df["Last_Update"]
#df["Date"] = df["Date"].astype('datetime64[ns]')
#df["Date"] = [time.date() for time in df["Date"]]
df["Horizontal"] = df["Latitude"] if df["Latitude"].notnull else df["Lat"]
df["Vertical"] = df["Longitude"] if df["Longitude"].notnull else df["Long_"]
print(df.columns)
print(df.dtypes)
df.head()
```

```python
download_df = df[[
    "Country",
    "State_Province",
    "FIPS",
    "Date",
    "Active",
    "Confirmed",
    "Deaths",
    "Recovered",
    "Incidence_Rate",
    "Case-Fatality_Ratio",
    "Horizontal",
    "Vertical",
    "Updated"
]].copy
download_df.to_csv(r'C:\Users\janin\Downloads\johns_hopkins.csv', index = False)
```
